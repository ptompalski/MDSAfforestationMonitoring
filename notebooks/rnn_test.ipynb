```python
import pickle
import pandas as pd
```

### With Site Features
Input Size = 12 (All satellite features : VIs, log time delta, cosine-transformed DOY)

Hidden Size = 16

Number of RNN layers = 1

Site Features = 4 (All site features : Age, Density, Type)

Epochs = 20

Learning Rate = 0.01

Validation Loss : 200-300

#### LSTM

```python
metrics = []
for threshold in [50.0, 60.0, 70.0]:
    with open(f'../results/eval_LSTM_{threshold}_site_feats.pkl', 'rb') as f:
        data = pickle.load(f)
        metrics.append(data['error_metrics_overall'])
df = pd.DataFrame(metrics, index=['50%', '60%', '70%'])
display(df)
print('Threshold :', threshold, '%')
display(data['conf_matrix_overall'])
display(data['error_metrics_age'].set_index('Age'))
```
